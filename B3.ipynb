```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('climate_data.csv')  #https://www.kaggle.com/datasets/mastmustu/weather-analysis
```

```python
avg_temp = df['Average temperature (°F)'].mean()
avg_dew_point = df['Average dewpoint (°F)'].mean()
```
